Для інтерполяції використаємо інтерполяційний поліном Ньютона, попередньо розбивши відрізок інтерполяції на 15 вузлів

```python
import numpy as np
import pandas
from scipy.misc import derivative
from scipy.linalg import solve_banded
from functools import reduce
import matplotlib.pyplot as plt

def getPoints(interval, n):
    return [interval[0] + i * (interval[1] - interval[0]) / n for i in range (0, n+1)]

def printPolynome(coeffs, n, points):
    factors = lambda pnts: ''.join([f'(x - {round(x_i, 5)})' for x_i in pnts]) if len(pnts) != 0 else ''
    polynome = f"{round(coeffs[0], 5)} + " + ' + '.join([f"{round(coeffs[i], 5)}*{factors(points[:i])}" for i in range(1, n)])
    print("Інтерполяційний поліном Ньютона P(x) =", polynome, '\n')
    
def newtonInterpolation(f, interval, n):
    points = getPoints(interval, n)
    func_vals = list(map(f, points))
    table = list(map(lambda x: [round(x[0], 4), round(x[1], 3)], zip(points, func_vals)))
    print("Таблиця значень функції у вузлах")
    print(pandas.DataFrame(table, columns=['x', 'f(x)']), '\n')
    def diffs(args):
        if len(args) == 1:
            return f(args[0])
        return (diffs(args[:-1]) - diffs(args[1:])) / (args[0] - args[-1])
    coeffs = [diffs(points[:i]) for i in range(1, n+1)]
    factors = lambda x, pnts: np.prod([x - x_i for x_i in pnts]) if len(pnts) != 0 else 1
    polynome = lambda x: sum([coeffs[i]*factors(x, points[:i]) for i in range(0, n)])
    printPolynome(coeffs, n, points)
    return polynome

def findC(h, n, func_vals):
    banded_mtrx = np.ndarray((3, n))
    banded_mtrx[1] = np.array([1.0] + [4*h for i in range(1, n-1)] + [1.0])
    banded_mtrx[0] =  np.array([0] + [h for i in range(1, n)])
    banded_mtrx[2] = np.array([h for i in range(1, n)] + [0])
    vec = np.array([0] + [3/h*(func_vals[i] - 2*func_vals[i-1] + func_vals[i-2]) for i in range(2, n)] + [0])
    return solve_banded((1, 1), banded_mtrx, vec)
    
def splineInterpol(f, interval, n):
    points = getPoints(interval, n)
    y = list(map(f, points))
    h = (interval[1] - interval[0]) / n
    a = y[:-1]
    c = findC(h, n, y)
    b = [(y[i+1] - y[i]) / h - h/3*(c[i+1] + 2*c[i]) for i in range(0, n-1)] + [(y[n] - y[n-1]) / h - 2*h/3*c[n-1]]
    d = [(c[i+1] - c[i])/(3*h) for i in range(0, n-1)] + [-c[n-1]/(3*h)]
    def interpolFunc(x):
        for i in range(0, n):
            if x < points[i+1] and x >= points[i]:
                return a[i] + b[i]*(x - points[i]) + c[i] * (x - points[i])**2 + d[i]*(x - points[i])**3
        if x == points[n]:
            return y[n]
    return interpolFunc

def findMax(func, interval):
    points = getPoints(interval, 300)
    vals = list(map(func, points))
    return max(vals)
```

```python
f = lambda x: np.square(x)*np.cos(x)
interval = [-np.pi / 2, np.pi]
n = 15

P = newtonInterpolation(f, interval, n)
epsilonP = lambda x: np.absolute(f(x) - P(x))
print("Максимальна похибkа інтерполяційного поліному Ньютона приблизно:", findMax(epsilonP, interval), '\n')

S = splineInterpol(f, interval, n)
epsilonS = lambda x: np.absolute(f(x) - S(x))
print("Максимальна похибkа сплайн-iнтерполяції приблизно:", findMax(epsilonS, interval))
theor_eps = 5/2*((interval[1] - interval[0]) / n)**3*findMax(lambda x: derivative(f, x), interval)
print("Теоретична похибkа сплайн-iнтерполяції:", theor_eps, '\n')


plt_points = getPoints(interval, 200)
plt_vuzols = getPoints(interval, n)

plt_P = list(map(P, plt_points))
plt_S = list(map(S, plt_points))
fig, ax = plt.subplots(2, 1, figsize=(17,17))
ax[0].plot(plt_points, plt_P)
ax[0].scatter(plt_vuzols, list(map(P, plt_vuzols)), color='red', label="вузли інтерполяції")
ax[0].set_title("Графік інтерполяційного поліному Ньютона при кількості вузлів n = 15")
ax[0].legend()
ax[1].plot(plt_points, plt_S)
ax[1].scatter(plt_vuzols, list(map(S, plt_vuzols)), color='red', label="вузли інтерполяції")
ax[1].set_title("Графік сплайн-iнтерполяції при кількості вузлів n = 15")
ax[1].legend()
plt.show()

print('\n')

plt_Perrs = list(map(epsilon, plt_points))
plt_Serrs = list(map(epsilonS, plt_points))
fig, ax = plt.subplots(2, 1, figsize=(17,17))
ax[0].plot(plt_points, plt_Perrs)
ax[0].set_title("Похибkа інтерполяційного поліному Ньютона при кількості вузлів n = 15")
ax[1].plot(plt_points, plt_Serrs)
ax[1].set_title("Похибkа сплайн-iнтерполяції при кількості вузлів n = 15")
plt.show()
```
